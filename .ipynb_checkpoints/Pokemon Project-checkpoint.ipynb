```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from numpy import array
from sklearn.preprocessing import StandardScaler
from matplotlib import pyplot as plt
from sklearn.linear_model import LogisticRegression
import numpy as np
```

```python
#Loading in the data
df = pd.read_csv('pokemon_alopez247.csv')
```

# Type Prediction Test (Nearest Neighbors)

```python
#Arranging data for type prediction test via Nearest Neighbor
types = df['Type_1'].unique()
df2 = df[:]
y = 0
for x in df2.index:
    if df2["Type_2"][x] in types:
        y += 1
        df2.loc[720 + y] = df2.loc[x]
        df2.loc[720 + y, 'Type_1'] = df.iloc[x][3]
df2 = df2.drop(columns = ['Type_2'])
df2 = df2.rename(columns = {'Type_1':'Type'})
df2 = df2.reset_index()
df2 = df2.drop(columns = ['index'])
```

```python
#Creating dataframe for first 720 pokemon type prediction test
df3 = df2[:720]
```

```python
#Grabbing numerical data
data = []
for x in range(720):
    z = []
    z.append(df3['Total'][x])
    z.append(df3['HP'][x])
    z.append(df3['Attack'][x])
    z.append(df3['Defense'][x])
    z.append(df3['Sp_Atk'][x])
    z.append(df3['Sp_Def'][x])
    z.append(df3['Speed'][x])
    z.append(df3['Height_m'][x])
    z.append(df3['Weight_kg'][x])
    z.append(df3['Catch_Rate'][x])
    data.append(z)
targets = []
for x in range(720):
    targets.append(df3['Type'][x])
```

```python
#Applying Nearest Neighbors
targets = array(targets)
data = array(data)
```

```python
x_train, x_test, y_train, y_test = train_test_split(data, targets, random_state = 42, stratify = targets)
```

```python
x_scaler = StandardScaler().fit(x_train)
```

```python
x_train_scaled = x_scaler.transform(x_train)
x_test_scaled = x_scaler.transform(x_test)
```

```python
train_scores = []
test_scores = []
for k in range(1, 20):
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(x_train_scaled, y_train)
    train_score = knn.score(x_train_scaled, y_train)
    test_score = knn.score(x_test_scaled, y_test)
    train_scores.append(train_score)
    test_scores.append(test_score)
    print(f"k: {k}, Train/Test Score: {train_score:.3f}/{test_score:.3f}")
    
    
plt.plot(range(1, 20), train_scores, marker='o')
plt.plot(range(1, 20), test_scores, marker="x")
plt.xlabel("k neighbors")
plt.ylabel("Testing accuracy Score")
plt.show()
```

```python
model_dict = {}
missing_pokemon = []
training_scores = []
testing_scores = []
never_wins = []
always_wins = []
h = 0
for z in df_800_pokemon.index:
    if len(battle_dict[z+1][0]) > 0:
        if 0 in list(battle_dict[z + 1][1]) and 1 in list(battle_dict[z + 1][1]):
            x = battle_dict[z + 1][0]
            y = battle_dict[z + 1][1]

            x_train, x_test, y_train, y_test = train_test_split(x, y, random_state=1, stratify=y)
            classifier = LogisticRegression()
            classifier.fit(x_train, y_train)

            training_scores.append(classifier.score(x_train, y_train))
            testing_scores.append(classifier.score(x_test, y_test))

            model_dict[z] = classifier
        else:
            if 1 in battle_dict[z + 1][1]:
                always_wins.append(df_800_pokemon.loc[z]["Name"]) #Not enough data
            else:
                never_wins.append(df_800_pokemon.loc[z]["Name"])
    else:
        missing_pokemon.append(df_800_pokemon.loc[z]["Name"])
    h += 1
    print(h)
```
